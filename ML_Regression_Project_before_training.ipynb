# Create a train set and a test set

<u>Notation</u>: DF - DataFrame<br>
### Divide the DF into 2 DF-s, using <font color='blue'>stratified sampling</font>:
- for training (<font color='blue'><b>strat_train_set</b></font>)
- for testing (<font color='blue'><b>strat_test_set</b></font>)

<u>Note</u>: use <i><b>median_income</b></i> for stratified sampling. Reason: the value of a house that a family can afford depends on the family's income.

### Pay attention to the following!!
- Both train and test sets should include <u>randomly selected instances</u>;<br>
- The algorithm will run many times before being ready to deploy in production => the testing set should be the same each time (do not generate a different test set each time, or the training algorithm will see all the dataset
  over time);
- Each record (fiecare rand din tabel) should have a unique identifier (eg., the ID could be a combination of `latitude` and `longitude`);
- The test set should be representative for all the population => this is why we use stratified sampling.

#### Definition:
<font color='blue'><i>strata:</i></font> each homogenous sugrup of the population.<br>

Because `median_income` is very important in deciding house prices (`median_house_value`) => strata-s will be categories of income (from very low to very high income)

#### Important!
Each income category (each strata) <u>should have a sufficient number of instances</u> => we should not have too many small strata (each strata should be large enough); otherwise, some income categories will have an overestimated importance.

### Procedure:
1. add a new column <i><b>income_cat</b></i> (categoria de venituri), with 5 categories : [0, 1.5), [1.5, 3), [3, 4.5), [4.5, 6), [6, inf)
(in NumPy, <i>inf</i> comes from <i>infinity</i>);<br>

<br>

2. split the dataframe into ***train*** and ***test*** sets, using the above categories for stratified sampling;

3. delete the ***income_cat*** column.

#### Note: this section is very important for later discussion on cross-validation!!!

**1. Add the <i>income_cat</i> column:**

```python
import numpy as np
from sklearn.model_selection import train_test_split


# The following line splits the DF, but should only be used if we do not care about stratified sampling:
# train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)

# Use pd.cut to create new column ("income_cat") with 5 categories of income;
# Each district is placed in one of these 5 categories:

housing["income_cat"] = pd.cut(housing["median_income"],
                               bins=[0., 1.5, 3.0, 4.5, 6.0, np.inf],  # last bin: from 6 to infinity
                               labels=[1, 2, 3, 4, 5])                 # these are not labels from ML!
# housing.sample(10)
housing["income_cat"].hist(figsize=(8, 5))  # Plots a histogram

plt.show()
print(housing["income_cat"].value_counts())
```

**2. Split the dataframe:**

```python

from sklearn.model_selection import StratifiedShuffleSplit  # preserves the % of samples for each class
                                                            # (which is very important)

# Establish how to make the split (eg. test size is 20%, only one split, etc.): 

splitIndices = StratifiedShuffleSplit(n_splits=1,
                                      test_size=0.2,
                                      random_state=42)

# More on StratifiedShuffleSplit:
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.StratifiedShuffleSplit.html


for train_index, test_index in splitIndices.split(housing, housing["income_cat"]):
#     print(train_index, test_index)
    strat_train_set = housing.loc[train_index]  # Creates a stratified set for training
    strat_test_set = housing.loc[test_index]    # Creates a stratified set for testing

# strat_train_set.tail()
strat_test_set.tail(10)

# The last column from the DF below is "income_cat"
```

```python
# Check the proportions of each category in the entire DF (before split):

print(housing["income_cat"].value_counts()/len(housing)*100)  # same % in the entire set
print()
print(strat_train_set["income_cat"].value_counts()/len(strat_train_set)*100)
print()
print(strat_test_set["income_cat"].value_counts()/len(strat_test_set)*100)
```

Notice that the proportions are almost identical when we compare the testing set with the entire DF.

**3. Drop column income_cat (not necessary anymore):**

```python

strat_test_set.drop("income_cat", axis=1, inplace=True)

strat_train_set.drop("income_cat", axis=1, inplace=True)


# A much ore elegant piece of code:

# for each_set in (strat_test_set, strat_train_set):
#     each_set.drop("income_cat", axis=1, inplace=True)

strat_test_set.head()
```

### Find correlations between dataframe attributes (columns)
(see chapter "Regression" from the Numerical Methods lecture)<br><br>
Understanding the correlation between the features and the target variable (the label, which is housing_median_value) is essential. Features (longitude, latitude, median_age, etc.) that are strongly correlated with the target variable (the label) are often more informative and <u>can</u> contribute significantly to the predictive power of the model.<br>
However, high correlation between features and the target variable in the training set may lead to <u>overfitting</u>, where the model becomes too specialized to the training data and performs poorly on new, unseen data. Regularization techniques can help mitigate overfitting.<br><br>
High correlation between features in a dataset can sometimes lead to <u>multicollinearity</u> issues. This occurs when two or more features in the dataset are highly correlated to each other, making it difficult for the model to differentiate their individual effects on the target variable. In such cases, feature selection or dimensionality reduction techniques may be applied to address the issue and improve the model's performance.<br>

Here, we will replace the missing values in ***total_bedrooms*** with the median number of bedrooms, as follows:

```python

from sklearn.impute import SimpleImputer   # SimpleImputer - sklearn class for dealing with missing values


imputer = SimpleImputer(strategy="median") # other values: mean (default), most_frequent, constant (use fill_value)

# Create a copy of the DF without the text column (SimpleImputer doesn't work on text values):
housing_num = housing.drop("ocean_proximity", axis=1)

imputer.fit(housing_num)    # computes the median values, one for each column;
                            # (maybe some NEW data will have missing values in other columns, so our code
                            #  should be able to find the medians for those columns)

print(imputer.statistics_)  # the statistics_ variable stores the computed medians for each column
print(housing_num.columns)

# print(housing_num.median().values)  # this is another way to compute the median and show values

X = imputer.transform(housing_num)    # replaces NA values with medians; returns a numPy array instead of a DF

np.set_printoptions(suppress=True)    # Pentru a nu printa array-ul in format exponential
print(f"X:\n {X} \n") 

# Change back the array into a DF:
housing_tr = pd.DataFrame(X,
                          columns=housing_num.columns,  # Se precizeaza care trebuie sa fie denumirile coloanelor
                          index=housing_num.index)      # Se precizeaza coloana cu rol de index
housing_tr
```

Scikit-Learn’s API is very well designed. The main design principles are:
1. **Consistency**. All objects share a consistent and simple interface:
    - <u>Estimators</u>. Any object that can estimate some parameters based on a dataset is called an estimator (e.g., an imputer is an estimator). The estimation itself is performed by the ***fit()*** method, and it takes only a dataset as a parameter (or two for supervised learning algorithms; the second dataset contains the labels). Any other parameter needed to guide the estimation process is considered a hyperparameter (such as an imputer’s strategy), and it must be set as an instance variable (generally via a constructor parameter).<br><br>
    
    - <u>Transformers</u>. Some estimators (such as an imputer) can also transform a dataset; these are called transformers. Once again, the API is quite simple: the transformation is performed by the ***transform()*** method whose parameter is the dataset to be transformed. It returns the transformed dataset. This transformation generally relies on the learned parameters, as is the case for an imputer. All transformers also have a convenience method called ***fit_transform()*** that is equivalent to calling ***fit()*** and then ***transform()*** (but sometimes ***fit_transform()*** is optimized and runs much faster).<br><br>
    
    - <u>Predictors</u>. Some estimators are capable of making predictions given a dataset; they are called **predictors**. For example, the **LinearRegression** model is a predictor: it can predict a median value of the houses in a certain district, given the district’s attributes (location, total number of rooms, median house age, etc.). A predictor has a ***predict()*** method that takes a dataset of new instances and returns a dataset of corresponding predictions. It also has a ***score()*** method that measures the quality of the predictions given a test set (and the corresponding labels in the case of supervised learning algorithms).<br><br>
2. **Inspection**. All the estimator’s hyperparameters are accessible directly via public instance variables (e.g., ***imputer.strategy***), and all the estimator’s learned parameters (computed with <i>fit</i>) are also accessible via public instance variables with an underscore suffix (e.g., ***imputer.statistics_***).<br><br>
3. **Nonproliferation of classes**. Datasets are represented as NumPy arrays or SciPy sparse matrices, instead of homemade classes. Hyperparameters are just regular Python strings or numbers.<br><br>
4. **Composition**. Existing building blocks are reused as much as possible. For example, it is easy to create a **Pipeline** estimator from an certain sequence of transformers followed by a final estimator, as we will see in our example project.<br><br>
5. **Sensible defaults**. Scikit-Learn provides reasonable default values for most parameters, making it easy to create an initial working system quickly.

### 3. Convert the <i>ocean_proximity</i> attribute to a <font color='blue'>dummy attribute</font> (<font color='blue'>one-hot encoding</font>)
Originea termenului "one-hot" vine de la ideea că doar un singur "bit" (sau element) este "aprins" sau setat la 1 într-un vector binar. Practic, într-un vector "one-hot", doar una dintre poziții are valoarea 1, indicând astfel categoria sau valoarea asociată acelei poziții.
#### Notations:
- <font color='blue'><i><b>housing_cat</b></i></font> is a dataset (2D array) containing only the <i>ocean_proximity</i> attribute
- <font color='blue'><i><b>housing_cat_1hot</b></i></font>: a one-hot sparse matrix that stores only the location of nonzero-s

Example of conversion from string values to dummy values:

### 4. Feature scaling (scalarea valorilor atributelor la valori similare)
Use scalers to have the same scale for all attributes' values.<br>

Scalarea este o practică esențială în domeniul machine learning, având un impact semnificativ asupra performanței algoritmilor de învățare automată. Câteva motive cheie pentru care este recomandată scalarea:

**Optimizarea convergenței**: Unii algoritmi de învățare automată (de exemplu metoda gradientului), pot converge mai rapid atunci când caracteristicile sunt scalate.

**Prevenirea dominarii unei caracteristici asupra altor caracteristici**: În absența scalarii, o caracteristică cu valori mai mari poate influența modelul mai mult decat o caracteristica cu valori mai mici. Prin aplicarea scalarii, se echilibrează contribuția fiecărei caracteristici in timpul procesului de antrenare.

**Îmbunătățirea performanței in cazul anumitor algoritmi**: Anumiti algoritmi ML sunt sensibili la scala (valoarea) caracteristicilor. Scalarea poate conduce la o îmbunătățire semnificativă a performanței acestor algoritmi.<br>

#### <font color='blue'>Normalization</font> (or <font color='blue'>Min-max scaling</font>):
$$X_{normalized}=\frac{X-X_{min}}{X_{max}-X_{min}}$$

- normalized values are between 0 and 1;
- good for neural networks, but affected by outliers;
- in Scikit-Learn, use the <font color='blue'><i>MinMaxScaler</i></font> transformer (***feature_range*** hyperparameter allows to change the default range (0-1) to another range).

Example of normalizaton:
- initial set: 2,3, 6, 7, 10
- normalize 7: (7-2)/(10-2) = 0.625

#### <font color='blue'>Standardization</font>
$$x_{stand}=\frac{x-\overline{x}}{S}$$<br>
**Note**: think of standard deviation (***S***) as an "average distance" (take the distance from each value to the mean, and compute the average of these distances using the following equation): $$S=\sqrt{\frac{\sum_{1}^{n}(x_i-\overline{x})^2}{N-1}}$$<br>
where ***N*** is the number of values in the sample, and $\overline{x}$ is the mean value of the sample.
- Standardization is less prone to outliers (este mai putin sensibila la valorile extreme);
- Standardization does not bound to a specific range (such as 0-1 in case of normalization); for some algorithms (like neural networks), this could be a problem; 
- In Scikit-Learn, use the <font color='blue'><i>StandardScaler</i></font> transformer.

#### Important note!
When training the scalers using the ***fit()*** function, use <u>only the training set (do NOT use the test set)</u>! After that, use the trained scaler to scale (***transform()*** function) the training set, test set, and also the new data.

***Creating a transformer class*** for all previous attributes discussed earlier; the class will add 3 more attributes:<br> <i>rooms_per_household, population_per_hosehold, bedrooms_per_room</i><br><br>
Methods that should be implemented (according to scikit-learn rules):<br>
<i>fit(), transform(), fit_transform()</i>
- <i>fit()</i> should return <i>self</i>;
- <i>fit_transform()</i> is included automatically by adding the ***TransformerMixin*** base class.
<br><br>

The class:
- will have one hyperparameter: ***add_bedrooms_per_room***. Two combinations are allowed (True or False), to test if this attribute improves or not the accuracy of the result.
- will return an array (NOT a DataFrame!).

```python
from sklearn.base import BaseEstimator, TransformerMixin  # BaseEstimator adds get_params() and set_params()
                                         # useful for hyperparameter tuning; do not use *args and **kwargs into
                                         # the __init__ function; TransformerMixin adds fit_transform()

rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6  # Column indexes (cols 3, 4, 5, 6 from array X, 
                                                                  # which is the NumPy array obtained after
                                                                  # replacing the NA values with the median value)

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):   # inherits from BaseEstimator & TransformerMixin
    def __init__(self, add_bedrooms_per_room=True):               # no *args and **kwargs here!
        self.add_bedrooms_per_room = add_bedrooms_per_room        # This is a hyperparameter
    
    # Here, the fit() method will not be used (objects generated by this class will not be trained on any dataset)
    # However, being a transformer class, the fit() method should exist, even if it only returns the object itself:
    
    def fit(self, X, y=None):
        return self  # this means that nothing happens with the object (the object is not trained),
                     # when the fit() method is called
        
    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]   # (All X rows, column 3)/(All X rows, col. 6)
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            
            # Concatenate new columns to the right side of matrix X, using the c_ function: 
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]


# Test how the class performs:

attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=True)

housing_extra_attribs = attr_adder.transform(housing.values)  # housing.values returns a NumPy representation of a
                                                        # DataFrame; use print(housing.values) to see the result
# print(housing, '\n')
# print(housing.values, '\n\n')
print(housing_extra_attribs[:, -4::])  # Printing only the last 4 columns
```

#### Procedure:
- Use a <font color='blue'><i><b>pipeline</b></i></font> to fill missing values, add atributes, and scale ONLY the numerical attributes;

The following cell uses the above transformer class.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler


num_pipeline = Pipeline([                          # List of tuples
    
    ("imputer", SimpleImputer(strategy="median")), # First, fill the missing values with median;
    
    ("attribs_adder", CombinedAttributesAdder()),  # Then, define and add new attributes
    
    ("std_scaler", StandardScaler())])  # Finally, scale all numerical values using standardization, in order
                                        # to bring the values in all columns in the same range (ML doesn't like
                                        # values that are too different from one another)
                                        # "imputer", "attribs_adder" & "std_scaler" - names we give, that can
                                        # be used later for hyperparameter tuning.

housing_num_tr = num_pipeline.fit_transform(housing_num)  # Transforms housing_num (which is a DataFrame without
                                                          # the "ocean_proximity" column) and returns an array with
print(housing_num.head(), '\n')                           # normalized values (housing_num_tr).

print(housing_num_tr[:6, :].round(2))
```

**Note:** All the estimators included within the pipeline must be transformers. The pipeline has the same methods like the last estimator. Following, the last one is <i>OneHotEncoder</i>, so the pipeline will have the <i>transform()</i> and <i>fit_transform()</i> methods.

```python
# Transforming numerical and categorical columns in a single pipeline, using ColumnTransform

from sklearn.compose import ColumnTransformer


# First, separate numerical from text attributes:
num_attribs = list(housing_num)    # List of column names (columns with numerical values, without "ocean_prox")
cat_attribs = ["ocean_proximity"]  # List with a single column name (column with text values)

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),      # This line applies previous num_pipeline on all numerical columns
    ("cat", OneHotEncoder(), cat_attribs)])  # This line transforms ocean_proximity values in dummy variables
                                             # "num" & "cat" are names we give, that can be used later for 
                                             # hyperparameter tuning
        
# Prepare the training model (housing) for final training:
housing_prepared = full_pipeline.fit_transform(housing)  # remember that "housing" is the training DF with all
                                                         # columns (numerical and text), but without labels.

# print the resulting array on the full width of the page in Jupyter
np.set_printoptions(edgeitems=10,   # Nr of array items in summary at beginning and end of each dimension (def: 3)
                    linewidth=600,  # Nr of characters per line for the purpose of inserting line breaks (def: 75)
                    precision=3,    # Number of digits of precision for floating point output (default 8)
                    suppress=True)  # Prints small floating point values using nonscientific notation (def: False)

print(housing_prepared[:6, :])

# NOTE: OneHotEncoder returns a sparse matrix, & num_pipeline returns a dense matrix => full_pipeline decides if
# ----  the resulting matrix will be sparse or dense depending on the ratio of non-zero cells (by default, the
#       ratio threshold is sparse_threshold=0.3 => sparse matrix if % of nonzero values is under 30%)

# Reset printoptions to default:
np.set_printoptions(edgeitems=3, infstr='inf', linewidth=75, nanstr='nan', precision=8,
                    suppress=False, threshold=1000, formatter=None)

# set_printoptions help page:
# https://numpy.org/doc/stable/reference/generated/numpy.set_printoptions.html
```
